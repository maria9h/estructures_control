### Exercici 1

Crea una funció que donat un Array d’una dimensió, et faci un resum estadístic bàsic de les dades. Si detecta que l’array té més d’una dimensió, ha de mostrar un missatge d’error.

```python
import numpy as np
import pandas as pd

a = np.array([1,2,3,4])

def resum(a):
    if a.ndim > 1:
        print('Error, l\'array té més d\' una dimensió!')

    else:
        s = pd.Series(a)
        print(s.describe())

resum(a)
```

### Exercici 2
Crea una funció que et generi un quadrat NxN de nombres aleatoris entre el 0 i el 100.

```python
import random

def quadrat():
    arr=[]
    n = random.randint(1,100)
    rows = n
    cols = n
    for i in range(rows):
        col = []
        for j in range(cols):
            col.append(random.randint(0,100))
        arr.append(col)
    print(arr)

quadrat()
```

### Exercici 3

Crea una funció que donada una taula de dues dimensions (NxM), et calculi els totals per fila i els totals per columna.

```python
import numpy as np

df = np.array([[1, 0, 0], [0, 2, 0], [0, 0, 3]])

def totals(a):
    total_f = np.sum(a, axis=1)
    print('Suma per files:', total_f)
    total_c = np.sum(a, axis=0)
    print('Suma per columnes:', total_c)

totals(df)
```

### Exercici 4
Implementa manualment una funció que calculi el coeficient de correlació. Informa-te'n sobre els seus usos i interpretació.

```python
import numpy as np
import math
df = np.array([[1, 2], [3, 4], [4, 5]])

def Peasron(a):
    xy = np.prod(a, axis = 1)
    sum_xy = np.sum(xy)
    x2 = a[:,0]**2
    sum_x2 = np.sum(x2)
    y2 = a[:,1]**2
    sum_y2 = np.sum(y2)
    mean = a.mean(axis = 0)
    cov = sum_xy/a.shape[0] - mean[0]*mean[1]
    cx = math.sqrt(sum_x2/a.shape[0] - mean[0])
    cy = math.sqrt(sum_y2/a.shape[0] - mean[0])
    r = cov/cx/cy
    print(r)
    
Peasron(df)
```
